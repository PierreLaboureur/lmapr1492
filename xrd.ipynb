```python
from pymatgen.io.cif import CifWriter
from pymatgen.ext.matproj import MPRester
import pymatgen.symmetry.analyzer as sa
from numpy import *
from pymatgen.core.operations import *
from pymatgen.analysis.diffraction import xrd
from pymatgen.analysis.diffraction.core import *
from matplotlib.pyplot import *

# Structure for material id
m = MPRester("kqOngsaR68ix5AD46")
structure = m.get_structure_by_material_id("mp-4280")
# Write to a CIF.
w = CifWriter(structure,0.1)
w.write_file('GaCuO2_mp-4280_symmetrized.cif')

A        = xrd.XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
patternA = A.get_pattern(structure, scaled=True, two_theta_range=(0, 90))
A.get_plot(structure, two_theta_range=(0, 90), annotate_peaks=True, ax=None, with_labels=True, fontsize=10)
hkls   = patternA.hkls
angle  = patternA.x
intens = patternA.y

print("On peut, avec les méthodes des classes XRDCalculator et AbstractDiffractionPatternCalculator, récupérer")
print("les indices de Miller des trois premiers pics du diffractogrammes pour un rayonnement Cu K alpha ainsi que")
print("l'intensité et l'angles correspondants à chacun de ces pics")
print("On peut vérifier l'exactitude de nos calculs via le diffractogramme")
print("premier pic :")
print(" - Angle : 2*theta  = " + str(angle[0]))
print(" - Intensité        = " + str(intens[0]))
print(" - Indice de Miller = " + str(hkls[0][0]['hkl']))
print("deuxième pic :")
print(" - Angle : 2*theta  = " + str(angle[1]))
print(" - Intensité        = " + str(intens[1]))
print(" - Indice de Miller = " + str(hkls[1][0]['hkl']))
print("troisième pic :")
print(" - Angle : 2*theta  = " + str(angle[2]))
print(" - Intensité        = " + str(intens[2]))
print(" - Indice de Miller = " + str(hkls[2][0]['hkl']))

```
